```python
import mapnik
import fiona
import json
```

```python
m = mapnik.Map(1400,1200)
m.background = mapnik.Color('gray')  
s = mapnik.Style() # style object to hold rules
r = mapnik.Rule() # rule object to hold symbolizers
# to fill a polygon we create a PolygonSymbolizer
polygon_symbolizer = mapnik.PolygonSymbolizer()
polygon_symbolizer.fill = mapnik.Color('#f2eff9')
r.symbols.append(polygon_symbolizer) # add the symbolizer to the rule object
r.filter = mapnik.Expression("[party] = '(R)'")
# to add outlines to a polygon we create a LineSymbolizer
line_symbolizer = mapnik.LineSymbolizer()
line_symbolizer.stroke = mapnik.Color('rgb(50%,50%,50%)')
line_symbolizer.stroke_width = 0.5
r.symbols.append(line_symbolizer) # add the symbolizer to the rule object
s.rules.append(r) # now add the rule to the style and we're done

s2 = mapnik.Style() # style object to hold rules
r2 = mapnik.Rule() # rule object to hold symbolizers
# to fill a polygon we create a PolygonSymbolizer
polygon_symbolizer2 = mapnik.PolygonSymbolizer()
polygon_symbolizer2.fill = mapnik.Color('#000000')
r2.symbols.append(polygon_symbolizer2) # add the symbolizer to the rule object
r2.filter = mapnik.Expression("[party] = '(D)'")
# to add outlines to a polygon we create a LineSymbolizer
r2.symbols.append(line_symbolizer) # add the symbolizer to the rule object
s2.rules.append(r2) # now add the rule to the style and we're done

m.append_style('R style',s)
m.append_style('D style',s2)# Styles are given names only as they are applied to the map


ds = mapnik.Shapefile(base='PLANH358',file='PLANH358')
layer = mapnik.Layer('Texas')
layer.datasource = ds

layer.styles.append('R style')
layer.styles.append('D style')
m.layers.append(layer)
m.zoom_all()

mapnik.render_to_file(m,'Texas.png', 'png')
```

```python
with open('texas.json', 'r') as df:
    data = json.load(df)

with fiona.open('PLANH358/PLANH358.shp', 'r') as one:
    schema2=one.schema.copy()
    schema2['properties']['party'] = 'str'
    print schema2
    with fiona.open('PLANH358/PLANH358.shp', 'w', crs=one.crs, driver=one.driver, schema=schema2) as source:
        for i in one:
            for j in data:
                if i['properties']['District'] == int(j['district']):
                    i['properties'].update(party=str(j['party']))
                    source.write(i)
                    
print 'done'
```
